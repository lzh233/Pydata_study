```python
#载入行星数据
planets = sns.load_dataset('planets')
print(planets.shape)
print(f"\n\n{planets.head()}")
```

```python
##分组 groupby, GroupBy对象 
#按照method分组
planets.groupby("method").sum().head()
```

```python
#将数据分组后，按method因子寻找orbital_period在各个因子下的中位数
planets.groupby("method")["orbital_period"].median()
```

```python
planets.groupby("method")["orbital_period"].describe()
```

```python
#GroupBy 对象的 aggregate()、 filter()、 transform()和 apply() 方法
# aggregate() 方法，
#构建数据
rng = np.random.RandomState(0)
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],'data1': range(6),'data2': rng.randint(0, 10, 6)},columns = ['key', 'data1', 'data2'])
df
```

```python
#(1) 累计 aggregate()实现更加复杂的累积操作比如字符串、 函数或者函数列表， 并且能一次性计算所有累计值。 
df.groupby("key").aggregate(['min','max',np.median,np.sum])
```

```python
#通过 Python 字典指定不同列需要累计的函数：
df.groupby("key").aggregate({'data1':"max", 
                             'data2':"min"})
```

```python
#(2) 过滤
#定义一个过滤函数
def filter_func(x):
    return x['data2'].std() > 4
df
```

```python
#只保留标准差大于4的组
print(f"{df.groupby('key').std()}\n\n\
{df.groupby('key').filter(filter_func)}")
```

```python
#(3) 转换。 
df.groupby('key').transform(lambda x: x - x.mean())
```
