```python
import numpy as np
import pandas as pd
```

```python
#pandas索引主要有iloc loc以及[]
#iloc ：基于位置的索引
#loc ：基于label的索引
```

```python
df=pd.DataFrame(np.random.rand(8,4),index=["a","b","c","d","e","f","g","h"],columns=["A","B","C","D"])
df
```

```python
###使用loc索引
df.loc["a",:]
```

```python
#同时取多行[[<要取的行名>]]
df.loc[["a","b","c"],:]
```

```python
#同时取多行(切片)[行名1:行名2]（与python原生切片不同不存在丢一行的情况）
df.loc["a":"c",:]
```

```python
#筛选数据A列数据大于0.5 B列数据大于0.4
df.loc[(df.A > 0.5) & (df.B > 0.4),:]
```

```python
#选择第A列大于0.5的CD列
df.loc[df.A > 0.5,["C","D"]]
```

```python
#对列进行索引
df.loc[:,["A","B","C"]]
```

```python
#对列进行切片
df.loc[:,"A":"C"]
```

```python
#行列同时
df.loc["a":"c","A":"C"]
```

```python
###使用iloc索引
#使用数字进行索引
df.iloc[0,:]
```

```python
#多行选取，使用list
df.iloc[[1,3,0],:]
```

```python
#也可以使用数组(小例子，随机取样两行)
df.iloc[np.random.randint(0,5,2),:]
```

```python
#iloc切片（存在丢一行的情况）
df.iloc[1:3,:]
```

```python
#可以实现对列的切片
df.iloc[:,1:4]
```

```python
df.iloc[1:3,1:4]
```

```python
#选取列
df.iloc[:,[1,2,0]]
```
